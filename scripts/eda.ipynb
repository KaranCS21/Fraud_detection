## Class Distribution
Credit card fraud detection is an imbalanced classification problem. Itâ€™s important to visualize how many fraudulent (1) vs. non-fraudulent (0) transactions are in the dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Check the class distribution (fraud vs non-fraud)
print("\nClass distribution:")
class_counts = df['Class'].value_counts()
print(class_counts)
```

```python
# Visualize the class distribution
sns.countplot(x='Class', data=df)
plt.title("Class Distribution: Fraud vs Non-Fraud")
plt.show()
```

## Correlation Matrix
itâ€™s helpful to visualize the correlation between features, especially if we plan to apply dimensionality
reduction techniques like PCA.

```python
# Visualizing the correlation matrix
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=False, cmap='coolwarm')
plt.title("Correlation Matrix of Features")
plt.show()
```
